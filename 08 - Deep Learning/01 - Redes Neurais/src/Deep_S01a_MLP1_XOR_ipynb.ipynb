```python
import numpy as np
from keras import layers, models
```

```python
X = np.array([[0 ,0],
              [0, 1],
              [1, 0],
              [1, 1]], dtype='float32')

y = np.array([[0],
              [1],
              [1],
              [0]], dtype='float32')
```

#**Construindo o Modelo**

image.png


```python
model = models.Sequential()

model.add(layers.Input(shape=(2,), name='input_layer'))

model.add(layers.Dense(8, activation = 'tanh', name = 'hidden1' ))

model.add(layers.Dense(8, activation = 'tanh', name = 'hidden2'))
```

```python
model.add(layers.Dense(1, activation = 'sigmoid', name = 'output' ))
```

```python
model.compile(optimizer = 'adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
model.fit(X, y, epochs=400, verbose=2)
```

```python
loss, accuracy = model.evaluate(X, y)

print(f'\nPerda final: {loss:.4f}')
print(f'Acurcia: {accuracy:.2f}')
```

```python
print('\nPrevises finais')

predictions = model.predict(X)
```

```python
for i in range(len(X)):
    input_pair = X[i]
    target = y[i][0]
    prediction = predictions[i][0]

    predicted_class = 1 if prediction > 0.5 else 0

    print(f'XOR ({int(input_pair[0])}, {int(input_pair[1])}) = {prediction:.4f} -> {predicted_class} (esperado: {int(target)} )')
```
